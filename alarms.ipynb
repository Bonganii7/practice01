```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Alarm_Details.csv", skiprows= 7)

print('Data is read in Pandas DataFrame!')
```

```python
df.head()
```

```python
alarm_counts = df['Alarm Type'].value_counts()


colors = ['blue', 'skyblue', 'purple', 'violet', 'pink', 'red','yellow']


plt.figure(figsize=(10, 6))
plt.bar(alarm_counts.index, alarm_counts.values, color=colors)


plt.xlabel('Alarm Type')
plt.ylabel('Alarm Count')
plt.title('Frequency of Different Alarm Types')


plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
